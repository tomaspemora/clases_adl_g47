```python
df = pd.read_csv('../Clase_2022_06_07/titanic.csv')
df = df.assign(
            cuadrado = lambda data: data.Age**2, 
            cuadrado_mas_10 = lambda data: data.cuadrado + 10,
            lower_embarked = lambda y: y.Embarked.str.lower(),
            cube_age = lambda x: x.cuadrado*x.Age
        )
```

Quiero sumar 10 a cada columna numerica en mi df

```python
%%time
data = df.select_dtypes(np.number)

for name, columna in data.iteritems():
    data[name] = columna + 10

data
```

Como podemos hacer funciones para pandas

```python
df = pd.read_csv('../Clase_2022_06_07/titanic.csv')
```

```python
# Quiero contar los nulos para ciertas variables

def contar_nulos(df, *columnas):
    columnas = list(columnas)
    return df[columnas].isnull().sum()

contar_nulos(df,'Embarked','Pclass', 'Age')
```

```python
df.describe().mean().max()
```

```python
df.pipe(contar_nulos, 'Embarked', 'Age') # condicion de pipe para que funcione es que el primer argumento debe ser un dataframe.
```

```python
(df.select_dtypes(np.number)
    .pipe(lambda x: x+10)
    .pipe(lambda x: x**2)
    .pipe(contar_nulos, 'Pclass', 'Age'))
```
