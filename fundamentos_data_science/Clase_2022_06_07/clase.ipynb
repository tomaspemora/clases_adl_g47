# Boolean Masks

```python
edad = int(input('Ingrese la edad: '))
edad > 30 # prueba lógica
```

```python
df.Age # Serie matematicamente esto es un vector.
```

```python
# python clasico se itera elemento por elemento
lista = [1,2,3]
for elemento in lista:
    print(elemento > 2)
```

```python
# en pandas las operaciones son vectorizadas
# en pandas nunca necesito un for loop ni un while (nunca ocupar iteritems o iterrows muy lentos)
# boolean mask
df.Age > 30
```

```python
# Se puede utilizar un boolean mask para poder filtrar

df[df.Age > 30] # esto me va a devolver los registros donde el boolean mask sea verdadero, es decir me devuelve las personas mayores de 30.as_integer_ratio
# Dicen que esta es la manera de filtrar en pandas, pero es añejo
```

```python
df[(df.Age > 30) & (df.Sex == 'female')] # & (vectorizado) y no and (no vectorizado)

# Pero... sintáxis poco amigable
```

```python
df.query('Age > 30 and Sex == "female" ') # no usar boolean masks para filtrar, mucho mejor query
# sobre 10000 filas query es mucho mas rápido
df.query('Survived == True and Pclass == 3 and Age < 18')
```

```python
df.query('Survived == True and Sex == "male" and Pclass == 1').shape # diferencias entre hombres y mujeres en primera clase
```

```python
df.query('Survived == True and Sex == "female" and Pclass == 1').shape
```
