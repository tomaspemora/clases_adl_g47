```python
# Modificar mi dataset... Transformaciones. VECTORIZADO
# Supongamos que queremos escribir todos los que tengan sexo 'female' con la primera mayúscula...

for row in df.Sex:
    if row == 'female':
        print(row.title())
    else:
        print(row)
```

```python
%%time
# Pero para guardar los cmabios necesitamos iterrows
for row, serie in df.iterrows():
    if serie.Sex == 'female':
        df.at[row,'Sex'] = serie.Sex.title()
```

```python
%%time
# Y esto es mucho mas rapido que iterrows.
df['Sex'] = [row.title() if row == 'female' else row for row in df.Sex]
```

```python
%%time
# Los obj str tienen el accesor str para aplicar funciones vectorizadas.
df['Sex'] = df.Sex.str.replace('female','Female')
```

```python
#df = pd.read_csv('../Clase_2022_06_07/titanic.csv', parse_dates=['Signing_date'])  # truco: parse_dates

# Que pasa si quiero crear muchas columnas nuevas, quiero año, el nombre del mes, el nombre del día

# df['year'] = df.Signing_date.dt.year # una copia
# df['month'] = df.Signing_date.dt.month_name()
# df['day'] = df.Signing_date.dt.day_name()

# esta manera evita tener que duplicar una columna entera (lo que puede ser costoso en memoria) y asigna nuevas columnas directamente. Tiene el problema de que estoy llamando al df más veces. Eso lo soluciono con lambda
df = df.assign(
    year = lambda x: x.Signing_date.dt.year,
    month = lambda x: x.Signing_date.dt.month_name(),
    day = lambda x: x.Signing_date.dt.day_name(),
    Cabin = 1 # reemplaza toda la columna con 1
    )
```

```python
df = pd.read_csv('../Clase_2022_06_07/titanic.csv', parse_dates=['Signing_date'])

# Tenemos 3 puertos: S(Inglaterra), C(Francia), Q(Irlanda)
# Supongamos que queremos que nuestro Embarked diga Inglaterra si es que el puerto S, y que siga diciendo lo mismo si es que es otro puerto

df.Embarked.mask(lambda x: x == 'S', 'Inglaterra') # if con else donde no modifico nada
```

```python
#Supongamos que quiero colocar sin cabina a los que no tienen cabina...
# where y mask
df.Cabin.mask(lambda x: x.isnull(), 'Sin cabina') # más natural, hace donde se cumple
df.Cabin.where(lambda x: x.notnull(), 'Sin cabina') # rara..., hace donde no se cumple
```

```python
df = pd.read_csv('../Clase_2022_06_07/titanic.csv', parse_dates=['Signing_date'])
# PANDAS PRIMERO !!! siempre q se pued
# Donde hay cabina, hay cabina y donde no hay cabina, le pongo no hay cabina
df.Cabin.mask(df.Cabin.notnull(), 'Hay cabina').mask(df.Cabin.isnull(), 'No hay cabina')
```

```python
# otra forma
df['Cabin_v2'] = np.where(df.Cabin.notnull(), 'Hay cabina', 'No hay cabina') # if vectorizado
```

```python
# Supongamos relabel diciendo si SibSp es > 0 casado y sino soltero. Y queremos indicar el sexo es decir... hombre casado, hombre soltero, mujer casada, mujer soltera

cond1 = (df.Sex == 'male') & (df.SibSp == 0) # hombre soltero 
cond2 = (df.Sex == 'male') & (df.SibSp > 0) # hombre casado
cond3 = (df.Sex == 'female') & (df.SibSp == 0) # mujer soltera
cond4 = (df.Sex == 'female') & (df.SibSp > 0) # mujer casada

cond_list = [cond1, cond2, cond3, cond4]
choice_list = ['hombre soltero', 'hombre casado', 'mujer soltera','mujer casada']
df['estado_civil'] = np.select(cond_list, choice_list) #if elif
df[:5]
```
