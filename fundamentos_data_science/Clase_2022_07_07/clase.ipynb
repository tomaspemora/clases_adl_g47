# Test de hipótesis

![](th.png)

Los test de una muestra pueden tener las siguientes hipótesis:
$$ H_a: \mu > \mu_0$$
$$ H_a : \mu < \mu_0$$
<center>one-tail</center>

$$ H_a: \mu \neq \mu_0$$
<center>two-tail</center>

p-value : (probabilistic value: no es una probabilidad....)

p_value > nivel de significancia .... falla en rechazar la hipótesis nula  
p_value < nivel de significancia .... rechaza la hipotesis nula

rechaza / falla en rechazar.... no tengo ningun antecendete que me permita pensar otra cosa....

Hay dos conceptos....

Hipotesis nula: $\mu_0 = \mu_1$  
Hipotesis alternativa: $\mu_0 \neq \mu_1$

$H_0$ y $H_1$ (tambien llamada $H_a$) son opuestas... por definición

![](ej.png)

H0: Consumo medio = 5.6 L  
H1: Consumo medio > 5.6 L  

Test de una cola...

DATO: confiabilidad + significancia = 1
significancia 1% equivalente a 99% confiabilidad

```python
df = pd.DataFrame({'consumo':[6.1,6.5,5.1,6,5.9,5.2,5.8,5.3,6.2,5.9,6.3]})
significancia = 0.01
df.mean()
```

```python
from scipy.stats import ttest_1samp, ttest_ind
statistic, pvalue = ttest_1samp(df['consumo'], 5.6, alternative = 'greater')
if pvalue > significancia:
    print('Fallo en rechazar la hipótesis nula')
else:
    print('Se rechaza la hipótesis nula')
```

Se rechaza la hipotesis nula, significa que el vehículo si tiene un rendimiento mayor a 5.6 L

## Problema 2 

Se dice que los habitantes de China tienen un peso promedio de 65 kg. Se toma una muestra para verificar si esto es cierto o no. Con una confianza del 95%

```python
valores = np.array([54.2, 66.4, 66.9 ,64.4, 71.1])
valores.mean()
```

```python
statistic, pvalue = ttest_1samp(valores, 65)
significancia = 1 - .95
if pvalue > significancia:
    print('Fallo en rechazar la hipótesis nula')
else:
    print('Se rechaza la hipótesis nula')
```

Esto significa que el promedio de peso de los habitantes de China es mayor a 65 kg.

## Problema 3
![](p3.png)

```python
ventas = pd.read_csv('ventas.csv')
ventas.mean()
```

```python
plt.figure(figsize=(10,8))
plt.hist(ventas.retail, alpha=.3, label='retail')
plt.hist(ventas.competencia, alpha=.3, label='competencia')
plt.title('Competencia de ventas')
plt.legend()
```

```python
ventas.sum()
```

Tipo de test debería hacer? Test de dos colas...


H0: Ambas poblaciones tienen el mismo promedio.  
H1: Ambas poblaciones tienen promedios distintos prom1 != prom2



```python
statistic, pvalue = ttest_ind(ventas.retail, ventas.competencia)
significancia = 0.05
if pvalue > significancia:
    print('Fallo en rechazar la hipótesis nula')
else:
    print('Se rechaza la hipótesis nula')
```

Significa que tanto el retail como la competencia vendieron lo mismo.
