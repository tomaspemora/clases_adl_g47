````python
import requests, json
import pprint

pp = pprint.PrettyPrinter(indent=4, compact=True, depth = 6).pprint

url = "https://jsonplaceholder.typicode.com/posts"

payload = {}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)

output = json.loads(response.text)
````

````python
pp(output)
````

````python
def requests_api(url, verb="GET"):
    payload = {}
    headers = {}
    response = requests.request(verb, url, headers=headers, data=payload)
    return json.loads(response.text)
````

````python
resultado = requests_api(url,'GET')
titulos = [elemento['title'] for elemento in resultado]
pp(titulos)
````

````python
# Cual es el mximo usedId
maximo_id = max(elemento['userId'] for elemento in resultado) # one liner del maximo
print(maximo_id)
````

````python
def promedio(lista):
    return sum(lista)/len(lista)

# Cual es el userId promedio
promedio_id = promedio([elemento['userId'] for elemento in resultado]) # promedio de los userId
print(promedio_id)
````

````python
# Cual es el id del titulo ms largo

largos_de_titulo = [(elemento['id'], len(elemento['title'])) for elemento in resultado]
maximo_largo = max(largos_de_titulo, key=lambda x: x[1])
print(maximo_largo)
````

````python
largos_de_titulo = [len(elemento['title']) for elemento in resultado]
maximo_largo = max(largos_de_titulo)
# maximo_largo = max(largos_de_titulo, key=lambda x: x[1])
id_max = [elements['id'] for elements in resultado if len(elements['title']) == maximo_largo]
print(resultado[id_max[0] - 1])
````

````python
api_fotos = requests_api('https://jsonplaceholder.typicode.com/photos','GET') # album de fotos...
````

````python
fotos = [foto['url'] for foto in api_fotos]
````

````python
def image_tag_creator(foto_url):
    return f"<img width='100' src = '{foto_url}' >\n"
def string_web_fotos(fotos):
    return ''.join(image_tag_creator(foto) for foto in fotos)
````

````python
def html_string(string_web_fotos):
    return  f"""
        <!DOCTYPE html >
        <html >
        <head >
        <meta charset = 'utf-8' >
        <meta name = 'viewport' content = 'width=device-width, initial-scale=1' >
        <title >Catalogo de imagenes inutiles</title >
        </head >
        <body >
        <h1>Catalogo de imagenes inutiles</h1>
        {string_web_fotos}
        </body >
        </html >
    """
````

````python
html_code = html_string(string_web_fotos(fotos[0:35]))
with open('fotos.html','w') as f:
    f.write(html_code)
````

````python
from IPython.display import display,HTML
display(HTML(html_code))
````

````python
import requests
import json


def requests_api_v2(url, verb, user_id=None, title=None, body=None):
    payload = f"""
                {{
                    "body": "{body}",
                    "title": "{title}",
                    "userId": {user_id}
                }}
                """
    headers = {
        'Content-Type': 'application/javascript'
    }

    response = requests.request(verb, url, headers=headers, data=payload)
    return json.loads(response.text)
print('GET')
pp(requests_api_v2(url, 'GET'))
print('POST')
pp(requests_api_v2(url, 'POST', 11, 'Post actualizado', 'Hola Hola'))
print('PUT')
pp(requests_api_v2(url+'/55', 'PUT', 11, 'Post actualizado', 'Hola Hola'))
````
